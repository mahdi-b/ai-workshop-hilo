````python
## The Prompting Skill You'll Need
Uncertain Content
1. Insufficient Context
Problem: Assuming the AI already knows project details.
 Examples:
"Fix this issue." → Which specific issue?
"Give me the results." → Which results, from what?
"I want that!" → What does that refer to?
"Add a function to calculate the metrics." → Which metrics?
Better examples:
"Give me the sales results from Q2."
"Fix the grammar and correct any typos in the introduction."
Solution: Dangling demonstratives are not your friend, and always provide full context—data structures, dependencies, and exact requirements. Be explicit about what to change or retrieve.
````

````python
2. Ambiguity (Vagueness or Conflicting Requirements)
Problem: Prompt is too vague or contains conflicting instructions.
 Examples:
"Analyze customer behavior." → Which customers, which behaviors, which timeframe?
"Improve the product." → Improve what—features, design, cost, or speed?
"Make it detailed but brief." → Contradiction.
"Be creative but follow this exact template." → Conflicting constraints.
Better examples:
"Identify customers who purchased 2+ times last quarter but not in the past 30 days."
"Suggest two UI changes to make the checkout process faster."
"Write a detailed analysis in exactly 200 words."
"Use this template structure but vary the language creatively."


Solution: Be specific, define boundaries, and clarify priorities when requirements conflict.
 Quick check:
Can I imagine the output in my mind’s eye?
Would a colleague know exactly what to deliver?
Do any requirements contradict each other?
````

````python
3. Silent Logical Errors
Problem: Prompt looks clear but hides an unstated rule, leading to misleading results.
Examples:
"Assign a risk score to every client." → Is 0 safe and 10 risky, or the reverse?
"List active users last week." → Does the week start Sunday or Monday?
Better examples:
"Risk score 0–10, where 0 = no risk and 10 = highest risk."
"List active users for Monday–Sunday, Aug 5–11."
Solution: Define rules and edge cases explicitly.


Quick check: Could two people interpret the request differently and both think they’re right? If yes, add a rule.
````

````python
3. Excessive/Irrelevant Context
Problem: Including unnecessary background information that obscures the actual request.
Examples:
"I've been working on this project for 3 months, my team consists of 5 people, we use agile methodology, our office is in Seattle, and we need to calculate the ROI." → Only the last part matters.
Better examples:
"Calculate ROI using revenue=$500k, costs=$300k."

Solution: Strip away backstory and context that doesn't directly affect the output. Lead with the ask.
    
Quick check: If you removed this sentence, would the output change? If no, delete it.
````

````python
Uncertain Form
4. Wrong Format
Problem: The format of the input or output is underspecified or assumed incorrectly.
 Examples:
Dates assumed MM-DD-YYYY but file uses DD-MM-YYYY.
"Give me a summary." → One sentence, three bullets, or a full page?
"Export the results." → Format not specified (CSV, JSON, Excel, text).
"Give me an image." → Format/size unclear (PNG/JPG/SVG, resolution?).


Better examples:
"Input dates are in MM-DD-YYYY format."
"Give me a three-bullet summary of the introduction."
"Export a CSV with columns: name, age, score."
"Generate a PNG at 800×600."


Solution: Always specify formats for inputs and outputs—column names, data types, units, length, style, or file formats.
````

````python
5. Compatibility Issues
Problem: Output is correct but incompatible with the required environment, style, or conventions.
 Examples:
Programming: Model uses pandas but project uses polars.
Writing: Revision doesn’t match Victorian prose style.
Design: Generated skeuomorphic 3D icons when flat icons were required.
Better examples:
"Use polars library version 2.1+ for data processing."
"Revise Chapter 3 in formal Victorian prose: ornate sentences, elaborate vocabulary, indirect dialogue, detailed descriptions."
"Generate flat icons using Material Design guidelines."
Solution: Anchor outputs to the tools, libraries, conventions, or styles already in use.


Practical check: What tools, styles, or formats already exist so you can reuse/follow them?
````

````python
### You can use AI as a tutor.

| | **AI as a teacher** | **AI as a worker** |
|---|---|---|
| When | You're in unfamiliar territory | You know what you need |
| Goal | Understand the problem | Solve the problem |
| Prompts | Open, exploratory: *"Help me understand..."* | Specific, constrained: PTCF framework |
| Risk if you skip | You spiral into compounding misunderstandings you can't detect | You get generic but fixable output |

Most people jump straight to "worker mode." The unlock is recognizing when you're in over your head and starting in "teacher mode" first — because **in an unfamiliar domain, a wrong answer you can't detect is far more dangerous than no answer at all.**
````

````python
### The fix — use AI to learn BEFORE you use AI to solve**

Before you ask AI to navigate an unfamiliar domain, use it to *teach you the terrain first*. Separate conversations. Different mindset.

**Step 1 — Learn the vocabulary.** *"I want to build a community garden on a vacant lot in Hilo. What are all the different types of permits and approvals that might be involved? Explain each one in plain language."* — Now "SMA" and "conservation district" aren't mystery acronyms. You know what they mean before you encounter them.

**Step 2 — Learn YOUR situation.** *"How do I find out what zoning my lot has and whether it's in a Special Management Area?"* — Now you can look it up and come back with actual facts, not guesses.

**Step 3 — Learn what to watch for.** *"What mistakes do people commonly make when navigating Hawai'i County permits for small community projects? Where do they get stuck?"* — Now you know the pitfalls before you fall into them.

**Step 4 — Learn to evaluate.** *"If I told you my lot is zoned [X] and is not in the SMA, which of those permits would actually apply to me?"* — NOW you're asking a specific, grounded question that the AI can answer accurately and that you can verify.

The first three steps might take 20 minutes. They save you from a week of confusion, wasted effort, or worse — acting on a plan that was wrong from the start.
````

````python
**Slide: "Workflow vs. Agent — Side by Side"**

| | **Workflow** (what we built) | **Agent** (what's arriving) |
|---|---|---|
| Who designs the steps? | You do | AI does |
| What happens if something unexpected comes up? | It stops or breaks | It adapts |
| Can it use multiple tools? | Only the ones you wired up | It picks the right tool for each step |
| Can it make judgment calls? | No — same path every time | Yes — that's the point |
| Risk | Low — predictable, auditable | Higher — you need to review the output |
| Best for | Repetitive, high-volume, same-every-time tasks | Complex, one-off, requires reasoning |
| Example | Every invoice → same 5 steps → spreadsheet | "Prepare me for Thursday's board meeting" |
````

````python
**Slide: "The Honest Caveat About Agents 1"**

*More autonomy = more power, but also more risk*

- Agents can make mistakes you don't catch — because you didn't design the steps, you might not know what it skipped
- They can take actions you didn't expect — if connected to email, an agent could *send* something, not just draft it
- They burn through context and usage limits faster — multi-step tasks are expensive
- **The supervision rule:** The more autonomy you give AI, the more carefully you need to review the output
- Right now: agents work best when a human reviews before anything goes external
- Think of it like a new hire — capable but needs oversight until you build trust
````

````python
**Slide: "The Honest Caveat About Agents 2"**

*Not everything called an "agent" is actually one*

**Most "agent platforms" are really just workflows with better marketing.**
- They wire up a fixed sequence of steps — AI reads a PDF, extracts data, writes to a sheet
- That's our invoice workflow. It's useful, but the AI isn't deciding anything. The steps are predetermined
- If you hear "agent" but the demo shows a rigid A → B → C pipeline, that's a workflow with an AI node in it. Nothing wrong with that — just know what you're buying

**There is a lot of snake oil in this space right now.**
- Hundreds of "AI agent" startups launched in the last year. Many are wrappers around the same models you already have access to
  - Solid promts, Custon wrappers around tools, cusotm pilelines, etc.
- If a platform can't clearly explain what it does that ChatGPT or Claude can't, be skeptical
- Red flags: vague demos, no free tier to test, promises of "autonomous business operations" with no mention of limitations
````

````python
**Slide: "The Honest Caveat About Agents 3"**

**Anything deployed in a real system needs verification, validation, and monitoring.**
- An agent running once in a demo is impressive. An agent running unsupervised 24/7 on your actual data is a different conversation
- You need logging — what did it do, when, and why?
- You need guardrails — what is it *not* allowed to do?
- You need a human review checkpoint before anything touches customers, finances, or legal obligations
- This is true for enterprise systems and it's true for your five-person nonprofit
````

````python
## 7. Privacy, Ethics, and the Honest Workforce Conversation (~8 min)

- **Data privacy — the rule of thumb your team needs**
  - Free-tier AI tools will most certainly use your inputs to train future models — check each tool's policy.
  - **Do not paste anything into a free AI tool that you wouldn't post on a public bulletin board** — no client data, student records, personnel files, health information, financial details
  - Paid/enterprise tiers (ChatGPT Team/Enterprise, Claude Pro, Google Workspace with Gemini) typically do not train on your data — but read the terms, they vary
  - Hawai'i context: orgs handling keiki data, health records, or Hawaiian cultural knowledge should be especially cautious — there is no "undo" once sensitive information enters a model's training data

- **Bias and cultural responsibility**
  - AI learns from internet data, which underrepresents and sometimes misrepresents Hawaiian, Pacific Islander, and rural communities
  - AI-generated images of "Hawaiian culture" routinely produce stereotypical, tourist-brochure imagery disconnected from lived reality — Hawai'i Business Magazine documented how models default to commercialized representations
  - AI-generated text about Hawaiian practices can flatten or misrepresent cultural meaning — always have cultural practitioners review before publishing
  - This is not a reason to avoid AI — it's a reason to use it critically and to ensure that local voices shape how these tools are applied

- **Workforce — saying the quiet part out loud**
  - AI will not eliminate most jobs. But people who know how to use AI will increasingly outperform those who don't — within the same roles
  - The real risk is not replacement; it's the growing productivity gap between AI-literate and AI-illiterate workers
  - The most valuable investment your organization can make right now: structured AI literacy for your entire team, not just the tech-savvy people
  - The right framing: "AI + human" workflows — AI handles the tedious drafting, formatting, summarizing; humans bring judgment, relationships, cultural knowledge, and accountability
````

````python
## 7. Where AI Will Waste Your Time or Get You in Trouble


- **Hallucinations are not a bug being fixed — they're a fundamental feature of how these models work**
  - The AI predicts probable words, not true ones. It will state fabricated facts with total confidence
  - An MIT study (Jan 2025) found AI models are 34% *more* likely to use words like "definitely" and "certainly" when generating *incorrect* information than when they're right
  - Real cases:
    - A New York lawyer was sanctioned after submitting a brief with AI-fabricated case citations (Mata v. Avianca, 2023) — now over 200 similar cases tracked globally in Charlotin's AI Hallucination Cases Database
    - Air Canada was ordered to honor a refund policy that was entirely invented by its customer service chatbot (2024)
    - Deloitte submitted a $440,000 report to the Australian government containing fake academic sources and a fabricated court quote (Oct 2025) — and a separate CA$1.6M report to Newfoundland with four nonexistent research papers (Nov 2025)
  - The practical rule: **never publish, submit, or act on AI output you haven't verified yourself**, especially for anything with legal, medical, or financial consequences

- **The use cases that reliably fail**
  - Anything requiring guaranteed factual accuracy without human review
  - Tasks that require your specific institutional knowledge, relationships, or cultural context
  - Nuanced cultural content — AI can misrepresent Hawaiian cultural values, practices, and language
  - High-stakes decisions where a wrong answer causes real harm

- **How to pick where to start — a simple filter:**
  - Is this task high-volume, repetitive, and low-risk if imperfect? → Great starting point
  - Could I catch an error easily if the AI makes one? → Proceed with review
  - If this output were 90-95% right, would that still save me significant time? → Worth trying
  - Is there a human review step before this reaches anyone? → Necessary safeguard
````
